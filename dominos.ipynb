```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example: assuming your DataFrame is named df

# Function to detect outliers using IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detect outliers for UnitPrice
unit_outliers, unit_lower, unit_upper = detect_outliers_iqr(df, 'unit_price')
print(f"UnitPrice outliers: {len(unit_outliers)}")
print(f"Lower bound: {unit_lower}, Upper bound: {unit_upper}")

# Detect outliers for TotalPrice
total_outliers, total_lower, total_upper = detect_outliers_iqr(df, 'total_price')
print(f"TotalPrice outliers: {len(total_outliers)}")
print(f"Lower bound: {total_lower}, Upper bound: {total_upper}")

# Plot boxplots
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(x=df['unit_price'])
plt.title("Boxplot - UnitPrice")

plt.subplot(1,2,2)
sns.boxplot(x=df['total_price'])
plt.title("Boxplot - TotalPrice")

plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_percentage_error
import numpy as np

mape_scores = {}

for variety in df['pizza_name_id'].unique():
    series = df[df['pizza_name_id'] == variety]['total_price'].reset_index(drop=True)

    # Split train/test
    train_size = int(len(series) * 0.8)
    train, test = series[:train_size], series[train_size:]

    try:
        # Fit ARIMA model
        model = ARIMA(train, order=(1,1,1))
        model_fit = model.fit()

        # Forecast
        forecast = model_fit.forecast(len(test))

        # Calculate MAPE
        mape = mean_absolute_percentage_error(test, forecast) * 100
        mape_scores[variety] = mape
        print(f"{variety} â†’ MAPE: {mape:.2f}")

    except:
        print(f"ARIMA failed for {variety}")

# Compute average MAPE across all pizza varieties
avg_mape = np.mean(list(mape_scores.values()))
print(f"\nOverall Average MAPE: {avg_mape:.2f}")
print(f"Overall ARIMA Accuracy: {100 - avg_mape:.2f}%")
```

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error
import pandas as pd

mape_scores = []

for variety in df['pizza_name_id'].unique():
    pizza_df = df[df['pizza_name_id'] == variety][['order_date', 'total_price']].rename(
        columns={'order_date': 'ds', 'total_price': 'y'}
    )

    # Ensure enough data
    if len(pizza_df) < 10:
        print(f"Skipping {variety} (not enough data)")
        continue

    # Train/test split
    train_size = int(len(pizza_df) * 0.8)
    train = pizza_df.iloc[:train_size]
    test = pizza_df.iloc[train_size:]

    if len(test) == 0 or len(train) == 0:
        print(f"Skipping {variety} (train/test split issue)")
        continue

    # Prophet model
    model = Prophet()
    model.fit(train)

    future = model.make_future_dataframe(periods=len(test), freq='D')
    forecast = model.predict(future)

    # âœ… Merge instead of loc to handle missing dates
    preds = forecast[['ds', 'yhat']].merge(test, on='ds', how='inner')

    if preds.empty:
        print(f"Skipping {variety} (no matching dates in forecast)")
        continue

    # Calculate MAPE
    mape = mean_absolute_percentage_error(preds['y'], preds['yhat'])
    mape_scores.append(mape)

    print(f"{variety} MAPE: {mape:.2%}")

# âœ… Average MAPE
if mape_scores:
    avg_mape = sum(mape_scores) / len(mape_scores)
    print(f"\nAverage Prophet MAPE across all pizza varieties: {avg_mape:.2%}")
else:
    print("âš ï¸ No valid pizza varieties for MAPE calculation.")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_percentage_error

# Ensure datetime type
df['order_date'] = pd.to_datetime(df['order_date'])

# âœ… Group by pizza_name_id
pizza_varieties = df['pizza_name_id'].unique()
mape_scores = {}

for pizza in pizza_varieties:
    data = df[df['pizza_name_id'] == pizza].groupby('order_date')['total_price'].sum().reset_index()
    data = data.set_index('order_date')

    if len(data) < 20:  # skip very small time series
        continue

    train_size = int(len(data) * 0.8)
    train, test = data.iloc[:train_size], data.iloc[train_size:]

    try:
        model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,7))
        results = model.fit(disp=False)

        forecast = results.forecast(steps=len(test))

        # âœ… Handle MAPE safely
        if test['total_price'].sum() == 0:
            mape = np.nan
        else:
            mape = mean_absolute_percentage_error(test, forecast)

        mape_scores[pizza] = mape * 100  # percentage
        print(f"{pizza} â†’ SARIMA MAPE: {mape*100:.2f}%")

    except Exception as e:
        print(f"Error for {pizza}: {e}")

# âœ… Average MAPE across all
valid_mapes = [m for m in mape_scores.values() if not np.isnan(m)]
if valid_mapes:
    avg_mape = np.mean(valid_mapes)
    print(f"\nâœ… Average SARIMA MAPE across all pizza_name_id: {avg_mape:.2f}%")
else:
    print("\nâš ï¸ No valid SARIMA MAPE scores available")

```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_percentage_error

# Ensure datetime
df['order_date'] = pd.to_datetime(df['order_date'])

pizza_varieties = df['pizza_name_id'].unique()
mape_scores = {}

for pizza in pizza_varieties:
    data = df[df['pizza_name_id'] == pizza].groupby('order_date')['total_price'].sum().reset_index()
    data = data.set_index('order_date')

    if len(data) < 20:
        continue

    # Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data[['total_price']])

    # Train-test split
    train_size = int(len(scaled_data) * 0.8)
    train, test = scaled_data[:train_size], scaled_data[train_size:]

    # Create sequences
    def create_sequences(dataset, time_step=5):
        X, y = [], []
        for i in range(len(dataset) - time_step):
            X.append(dataset[i:(i+time_step), 0])
            y.append(dataset[i+time_step, 0])
        return np.array(X), np.array(y)

    time_step = 5
    X_train, y_train = create_sequences(train, time_step)
    X_test, y_test = create_sequences(test, time_step)

    # Reshape for LSTM
    X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    # Build LSTM
    model = tf.keras.Sequential([
        tf.keras.layers.LSTM(50, return_sequences=True, input_shape=(time_step, 1)),
        tf.keras.layers.LSTM(50),
        tf.keras.layers.Dense(1)
    ])

    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=10, batch_size=16, verbose=0)

    # Forecast
    y_pred = model.predict(X_test)
    y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1))
    y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1, 1))

    # âœ… Safe MAPE
    if y_test_rescaled.sum() == 0:
        mape = np.nan
    else:
        mape = mean_absolute_percentage_error(y_test_rescaled, y_pred)

    mape_scores[pizza] = mape * 100
    print(f"{pizza} â†’ LSTM MAPE: {mape*100:.2f}%")

# âœ… Average MAPE
valid_mapes = [m for m in mape_scores.values() if not np.isnan(m)]
if valid_mapes:
    avg_mape = np.mean(valid_mapes)
    print(f"\nâœ… Average LSTM MAPE across all pizza_name_id: {avg_mape:.2f}%")
else:
    print("\nâš ï¸ No valid LSTM MAPE scores available")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# --- Ensure datetime in sales data ---
df['order_date'] = pd.to_datetime(df['order_date'])
df = df.sort_values('order_date')

# --- Aggregate daily total sales (quantity) ---
daily_sales = df.groupby('order_date')['quantity'].sum()

# --- Train ARIMA model ---
model = ARIMA(daily_sales, order=(5,1,0))   # you can tune (p,d,q)
model_fit = model.fit()

# --- Forecast next 7 days ---
forecast = model_fit.forecast(steps=7)

# Convert forecast to DataFrame
next_week = pd.DataFrame({
    'ds': pd.date_range(start=daily_sales.index.max() + pd.Timedelta(days=1), periods=7),
    'predicted_qty': forecast.values
})

# --- Allocate predicted sales across pizzas using historical proportions ---
pizza_share = df.groupby('pizza_name_id')['quantity'].sum()
pizza_share = pizza_share / pizza_share.sum()

pizza_forecast_list = []
for _, row in next_week.iterrows():
    for pizza_id, share in pizza_share.items():
        qty = row['predicted_qty'] * share
        pizza_forecast_list.append([row['ds'], pizza_id, qty])

pizza_forecast = pd.DataFrame(pizza_forecast_list, columns=['ds', 'pizza_name_id', 'predicted_qty'])

# --- Merge with pizza ingredients dataset ---
ingredients_df = pd.read_excel("Pizza_ingredients.xlsx", sheet_name="Pizza_ingredients")
merged = pizza_forecast.merge(ingredients_df, on='pizza_name_id')

# --- Calculate ingredient requirement ---
merged['ingredient_qty'] = merged['predicted_qty'] * merged['Items_Qty_In_Grams']

# --- Consolidated Purchase Order (weekly total) ---
purchase_order = (
    merged.groupby(['pizza_name_id', 'pizza_name', 'pizza_ingredients'])['ingredient_qty']
    .sum()
    .reset_index()
)

print("ðŸ“¦ Consolidated Purchase Order for Next Week:")
print(purchase_order)

# Save to Excel/CSV
purchase_order.to_excel("Consolidated_Purchase_Order_ARIMA.xlsx", index=False)

```
